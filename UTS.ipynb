```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import os
import joblib

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

membaca file/dataset menggunakan pandas

```python
df = pd.read_csv('Dataset_B_hotel.csv')
```

menunjukkan 5 baris teratas pada dataset

menunjukkan info (Jumlah data, data type, etc) dari setiap variable yang ada dalam dataset

Menunjukkan jumlah duplicate dan unique values

```python
print(f'Duplicate Values: {df.duplicated().sum()}')

print(f'\nUnique Values: \n{df.nunique()}')
```

Menghapus missing value pada datast

```python
df = df.dropna()
df = df.reset_index(drop=True)
```

menunjukkan info setelah menghapus missing value

```python
df.info()
```

Menghapus/drop column Booking_ID karena hanya berisikan ID unik yang tidak relevan untuk model

```python
df = df.drop(columns=['Booking_ID'], axis=1)
df
```

Melakukan label encoding pada variable target, dikarenakan hanya ada 2 values ('Canceled' atau 'Not_Canceled') maka label encoder sangat sesuai untuk problem tersebut

```python
encode = LabelEncoder()
df['booking_status'] = encode.fit_transform(df['booking_status'])
```

Check untuk sebelum membuat data input dan output, serta untuk menentukan berapa variabel feature_x (17) dan target_y (1)

```python
df.info()
```

```python
df
```

Membuat dataframe feature dan target untuk split

```python
feature = df.drop(columns='booking_status', axis=1)
target = df['booking_status']
```

Melakukan split data 80% Train dan 20% Test

```python
x_train, x_test, y_train, y_test = train_test_split(feature, target, test_size=0.2, random_state=42)
```

Melihat hasil split data

```python
x_train.info()
```

Melakukan encoding terhadap kolum kategorikal menggunakan OneHotEncode

```python
categorical_cols = ['type_of_meal_plan', 'room_type_reserved', 'market_segment_type']

# Inisialisasi encoder
OneEncode = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Fit dan transform hanya di data training
encoded_train = OneEncode.fit_transform(x_train[categorical_cols])
encoded_train_df = pd.DataFrame(
    encoded_train,
    columns=OneEncode.get_feature_names_out(categorical_cols),
    index=x_train.index  # Set index agar cocok saat concat
)

# Transform data testing (tanpa fit ulang)
encoded_test = OneEncode.transform(x_test[categorical_cols])
encoded_test_df = pd.DataFrame(
    encoded_test,
    columns=OneEncode.get_feature_names_out(categorical_cols),
    index=x_test.index  # Set index agar cocok saat concat
)

# Gabungkan dengan kolom numerik
x_train = pd.concat([x_train.drop(columns=categorical_cols), encoded_train_df], axis=1)
x_test = pd.concat([x_test.drop(columns=categorical_cols), encoded_test_df], axis=1)
```

Melihat hasil .info setelah encode

```python
x_train.info()
```

Melakukan reset index agar entries (jumlah kolom dan baris) sama dan teratur

```python
x_train = x_train.reset_index(drop=True)
x_test = x_test.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)
```

Melakukan model fitting menggunakan random forest

```python
RF_model = RandomForestClassifier(random_state=42)
RF_model.fit(x_train, y_train)
```

```python
RF_pred = RF_model.predict(x_test)
```

Menampilkan Classification Report model Random Forest

```python
print("Classification Report for Random Forest:")
print(classification_report(y_test, RF_pred, target_names=['Canceled', 'Not_Canceled']))
```

Melakukan model fitting menggunakan XGBoost

```python
XGB_model = XGBClassifier(random_state=42)
XGB_model.fit(x_train, y_train)
```

```python
XGB_pred = XGB_model.predict(x_test)
```

Menampilkan Classification Report model XGBoost

```python
print("Classification Report for XGBoost:")
print(classification_report(y_test, XGB_pred, target_names=['Canceled', 'Not_Canceled']))
```
